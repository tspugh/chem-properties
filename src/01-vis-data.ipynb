```python
# https://stackoverflow.com/questions/57062757/how-to-generate-a-graph-from-a-smiles-molecule-representation
from rdkit import Chem
smiles = full_train_data.loc[4836, "SMILES"]
print(smiles)
ex_mol = Chem.MolFromSmiles(smiles)
ex_mol
```

```python
# Rows in which at least one of the fields is NA
full_train_data[full_train_data.isna().any(axis=1)]
```

```python
# Count of NA values per column
full_train_data.apply(lambda value: value.isna()).sum(0)
```

```python
na_mask = full_train_data.isna()
valid_fields = na_mask.apply(lambda row: pandas.Series({
    'count': row.sum(),
    'columns': ', '.join(row.index[row].tolist())
}), axis=1)
grouped = valid_fields.groupby('columns').size().sort_values().reset_index(name='count')
grouped
```
